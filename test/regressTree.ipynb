```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

# Create a random dataset
rng = np.random.RandomState(1)
X = np.random.random([1000,800])
y = np.sin(np.sum(X,axis=1))

# y[::5] += 2 * (0.5 - rng.rand(32)) # 每五个点增加一次噪音
```

```python
tmp = np.sum(X,axis=1)
```

```python
print(X)
print(y)
```

```python
# Fit regression model
# regr_1 = DecisionTreeRegressor(max_depth=2)
# regr_2 = DecisionTreeRegressor(max_depth=5)
regr_3 = DecisionTreeRegressor(max_depth=8)
# regr_1.fit(X, y)
# regr_2.fit(X, y)
regr_3.fit(X, y)
```

```python
regr_3.predict(X)
```

```python
from IPython.display import Image  
from sklearn import tree
import pydotplus 

#regr_1 和regr_2
dot_data = tree.export_graphviz(regr_3, out_file=None,  #regr_1 是对应分类器
                         feature_names=[str(x) for x in list(np.arange(0,800))],   #对应特征的名字
                         class_names=['y'],    #对应类别的名字
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
graph.write_png('example.png')    #保存图像
Image(graph.create_png()) 

```
